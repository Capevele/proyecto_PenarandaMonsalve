```python
def seleccionaFilas(campo,numrecords):
  conn = sql.connect(base_de_datos)
  cursor=conn.cursor()
  query = f"SELECT * FROM TLT_propuesto_tabla ORDER by {campo} DESC LIMIT {numrecords};"
  cursor.execute(query)
  datos=cursor.fetchall() # para enviarlo a una variable de Python y poder visualizar
  conn.commit()
  for row in datos:
    print(row)
  cursor.close()
  conn.close()
```

```python
seleccionaFilas('OriginCity',10) # Seleccion e impresion de registros en la tabla TLT_propuesto_tabla
```

```python
# Muestra la informacion de las columnas en la tabla TLT_propuesto_tabla
conn=sql.connect(base_de_datos)
cursor=conn.cursor()
query=f"pragma table_info(TLT_propuesto_tabla);"
cursor.execute(query)
datos=cursor.fetchall()
for row in datos:
    print(row)
cursor.close()
conn.close()
```

```python
def sumarizaTLT(porcampo):
  numregistros=int(input("Ingrese el numero de lineas a visualizar: "))
  conn=sql.connect(base_de_datos)
  cursor=conn.cursor()
  query = f"""
  SELECT {porcampo}, SUM(TLT_propuesto)
  FROM TLT_propuesto_tabla
  GROUP BY {porcampo}
  LIMIT {numregistros}
  """
  cursor.execute(query)
  results=cursor.fetchall()
  for result in results:
    print(result)
  cursor.close()
  conn.close()
```

```python
sumarizaTLT('Source_Dest')
```

#CRUD Aplicado a una de las tablas de la base de datos.

Se escogera una de las tablas de dimension, la que contiene la informacion de las bodegas. Esto para no afectar el reporte final

```python
def imprimenrecords(nom_tabla,num_records):
  conn=sql.connect(base_de_datos)
  cursor=conn.cursor()
  query=f"SELECT * FROM {nom_tabla} LIMIT {num_records};"
  cursor.execute(query)
  datos=cursor.fetchall()
  for info in datos:
      print(info)
  cursor.close()
  conn.close()
```

```python
# Conexion a la tabla e impresion de los primeros 10 registros
imprimenrecords('Bodegas_Info_tabla',10)
```

```python
# Impresion de la informacion de las columnas que componen la tabla. El query presentado retornara:
# cid: El id de la columna
# name: El nombre de la columna
# type: El tipo de dato
# notnull: Si la columna es null
# dflt_value: El valor por defecto de la columna
# pk: Si la columna tiene una llave primaria
conn=sql.connect(base_de_datos)
cursor=conn.cursor()
query=f"pragma table_info(Bodegas_Info_tabla)" #
cursor.execute(query)
datos=cursor.fetchall()
for row in datos:
    print(row)
cursor.close()
conn.close()
```
